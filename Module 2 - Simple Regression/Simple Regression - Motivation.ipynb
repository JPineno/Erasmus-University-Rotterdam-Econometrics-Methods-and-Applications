```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import our dataset
expen = pd.read_excel('TrainExer11.xls')
expen.head()
```

```python
# set index to observation
expen = expen.set_index('Observation')
expen.head()
```

__(a)__

```python
# generate a histogram for the variable "age"
plt.hist(expen['Age'], edgecolor='black')

plt.title('Age Histogram')
plt.xlabel('Years')
plt.ylabel('Frequency')

plt.show()
```

```python
# generate a histogram for the variable "expenditures"
plt.hist(expen['Expenditures'], edgecolor='black')

plt.title('Average Daily Expenditures Histogram')
plt.xlabel('Monetary Units')
plt.ylabel('Frequency')

plt.show()
```

```python
# generate a scatter plot with expenditures on the
# y-axis, and age on the x-axis
plt.scatter(expen['Age'], expen['Expenditures'])

plt.title('Expenditures vs Age Scatter Plot')
plt.xlabel('Age (Years)')
plt.ylabel('Expenditures (Monetary Units)')

plt.show()
```
