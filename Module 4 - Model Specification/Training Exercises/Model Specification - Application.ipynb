```r
# import the stock index dataframe
stocks <- read_excel('TrainExer3-5.xlsx')
head(stocks)
```

__(a)__

```r
# regress "LogEqPrem" on a constant and all five explanatory varables, and look at R-squared
logeq5 <- lm(LogEqPrem ~ BookMarket + NTIS + EarnPrice + DivPrice + Inflation, data=stocks)
summary(logeq5)
```

```r
# regress "LogEqPrem" on a constant and "BookMarket", and look at R-squared
logeq_bm <- lm(LogEqPrem ~ BookMarket, data=stocks)
summary(logeq_bm)
```

__(b)__

```r
# store the fitted "LogEqPrem" based on the last regression in (a)
fitted_logeqp <- fitted(logeq_bm)
fitted_logeqp
```

```r
# add "fitted_logeqp" to the "stocks" dataframe
stocks <- stocks %>% mutate(fitted_logeqp)
stocks
```

```r
# regress "LogEqPremium" on a constant, "BookMarket", and the square of "fitted_logeqp"
reset <- lm(LogEqPrem ~ BookMarket + I(fitted_logeqp^2), data=stocks)
summary(reset)
```

```r
# to get the F-test on "fitted_logeqp", square the t-value, and see if we get 3.446
f_logeqp <- (summary(reset)$coefficients["I(fitted_logeqp^2)", "t value"])^2
f_logeqp
```

__(c)__

```r
# store the sum of square residuals of the last regression in (a)
ssr_logeqp <- sum((resid(logeq_bm))^2)
ssr_logeqp
```

```r
# regress "LogEqPrem" on a constant and "BookMarket", over the subsample 1927-1979
logeq_bm1 <- stocks %>% filter(between(Year, 1927, 1979)) %>% lm(LogEqPrem ~ BookMarket, data = .)
summary(logeq_bm1)
```

```r
# store the sum of square residuals of "logeq_bm1"
ssr_logeqp1 <- sum((resid(logeq_bm1))^2)
ssr_logeqp1
```

```r
# regress "LogEqPrem" on a constant and "BookMarket", over the subsample 1980-2013
logeq_bm2 <- stocks %>% filter(between(Year, 1980, 2013)) %>% lm(LogEqPrem ~ BookMarket, data = .)
summary(logeq_bm2)
```

```r
# store the sum of square residuals of "logeq_bm2"
ssr_logeqp2 <- sum((resid(logeq_bm2))^2)
ssr_logeqp2
```

```r
# get the rest of variables we need for computing the Chow Break Test
k = length(coef(logeq_bm))
n1 = length(stocks[['Year']][1927:1979])
n2 = length(stocks[['Year']][1980:2013])
```

F-test for the Chow break test:

$$F = \frac{(SSR_{\text{pooled}} - (SSR_1 + SSR_2)) \ \big/ \ k}{(SSR_1 + SSR_2) \ \big/ \ (n_1 + n_2 - 2k)}$$

```r
# compute the Chow break test following the formula above, and see if we get 2.269
cb_test = ((ssr_logeqp-(ssr_logeqp1+ssr_logeqp2))/k)/((ssr_logeqp1+ssr_logeqp2)/(n1+n2-2*k))
cb_test
```

__(d)__

F-test for the Chow forecast test:

$$F = \frac{(SSR_{\text{pooled}} - SSR_1) \ \big/ \ n_2}{SSR_1 \ \big/ \ (n_1 - k)}$$

```r
# compute the Chow forecast test following the formula above, and see if we get 0.765
cf_stat = ((ssr_logeqp-ssr_logeqp1)/n2)/(ssr_logeqp1/(n1-k))
cf_stat
```
