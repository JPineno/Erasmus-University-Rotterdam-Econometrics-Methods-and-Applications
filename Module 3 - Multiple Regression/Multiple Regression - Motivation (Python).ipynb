```python
# imports
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
# import our dataset
wage = pd.read_excel('TrainExer21.xls')
wage.head()
```

```python
# set index to observation
wage = wage.set_index('Observation')
wage.head()
```

__(a)__

```python
# define the variables
x = sm.add_constant(wage['Female'])
y = wage['LogWage']

# fit the model using OLS
log_wage = sm.OLS(y, x).fit()

# show summary of the model
log_wage.summary()
```

__(b)__

```python
# access the residuals of the model
resid = log_wage.resid
resid
```

```python
# regress e on a constant and eduction
ex = sm.add_constant(wage['Educ'])
ey = resid

# fit the model using OLS
e = sm.OLS(ey, ex).fit()

# show summary of the model
e.summary()
```

```python
# regress e on a constant and the part-time job dummy
ex = sm.add_constant(wage['Parttime'])
ey = resid

# fit the model using OLS
e = sm.OLS(ey, ex).fit()

# show summary of the model
e.summary()
```
